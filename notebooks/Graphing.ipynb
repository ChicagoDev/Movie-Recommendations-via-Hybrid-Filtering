```python
casting_df_copy = pd.read_csv('../data/the-movies-dataset/credits.csv')
```

```python
casting_df_copy.cast = casting_df_copy.cast.apply(lambda cast: ast.literal_eval(cast))
```

```python
dfs = []
for movie in casting_df_copy.itertuples():
    df = json_normalize(movie[1])
    movie_id = movie[3]
    df['movie'] = movie_id
    dfs.append(df)
```

```python
len(dfs)
```

```python
movie_casts = pd.concat(dfs)
```

```python
movie_casts.to_csv('../data/the-movies-dataset/cast_for_network.csv')
```

```python
import pickle
with open ('../data/the-movies-dataset/cast_for_network.pkl', 'wb') as fl:
    pickle.dump(movie_casts, fl)
```
